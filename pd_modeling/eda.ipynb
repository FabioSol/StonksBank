```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('../files/credit_risk_data_v2.csv', low_memory=False)
```

## General Eda

```python

def eda(df):
    # Get the columns by data type
    cat_cols = list(df.select_dtypes(include=['object']).columns)
    cont_cols = list(df.select_dtypes(include=['int', 'float']).columns)

    # Count the missing values for each column
    missing_counts = df.isnull().sum()

    # Print the missing value counts for categorical and continuous columns
    print("Missing Value Counts:")
    print("---------------------")
    print("Categorical:")
    print(df[cat_cols].isnull().sum())
    print("\nContinuous:")
    print(df[cont_cols].isnull().sum())

    # Print the range, mean, and standard deviation for continuous columns
    print("\nContinuous Column Statistics:")
    print("---------------------------------")
    for col in cont_cols:
        print(col)
        print("Range: [{}, {}]".format(df[col].min(), df[col].max()))
        print("Mean: {:.2f}".format(df[col].mean()))
        print("Standard Deviation: {:.2f}".format(df[col].std()))

        # Plot a histogram of the continuous column
        sns.histplot(df[col], kde=False, bins='auto')
        plt.show()

    # Print the counts for each category in categorical columns
    print("\nCategorical Column Value Counts:")
    print("---------------------------------")
    for col in cat_cols:
        print(col)
        print(df[col].value_counts(dropna=False))
        print()
```

```python
eda(dataset)
```
